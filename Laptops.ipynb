```python
import pandas as pd
def laptops(soup):
    prod_name = [i.text.strip() for i in soup.select('div._4rR01T')]
    price = [''.join((i.text.strip().split('₹')[1].split(','))) for i in soup.select('div._25b18c')]
    rating= [i.find_all('div')[0].text.strip() for i in soup.select('div.gUuXy-')]
    total_ratings= [''.join(i.text.strip().split(' ')[0].split(','))for i in soup.select('span._2_R_DZ')]
    total_reviews= [i.text.split('Ratings\xa0&\xa0')[1].split(' ')[0] for i in soup.select('span._2_R_DZ')]
    specifications=[i.text for i in soup.select('div.fMghEO')]
    return pd.DataFrame({'Product Name':prod_name,
                         'Price':price,
                         'Rating':rating,
                         'Number of ratings':total_ratings,
                         'Number of reviews':total_reviews,
                         'Laptop Specifications':specifications})
```

```python
df=laptops(soup)
df
```

```python
stud='https://www.flipkart.com/laptops/pr?sid=6bo%2Cb5g&fm=neo%2Fmerchandising&iid=M_8b3b3f65-7ceb-4375-912c-d2bcdde87c58_1_372UD5BXDFYS_MC.34WHNYFH5V2Y&otracker1=hp_rich_navigation_PINNED_neo%2Fmerchandising_NA_NAV_EXPANDABLE_navigationCard_cc_13_L1_view-all&cid=34WHNYFH5V2Y&p%5B%5D=facets.brand%255B%255D%3DAPPLE&otracker=clp_metro_expandable_6_26.metroExpandable.METRO_EXPANDABLE_Apple_laptops-store_SKIHMOPFPDC3_wp9&fm=neo%2Fmerchandising&iid=M_b4d0e2ef-aaac-4e7e-9272-9b8be53c2dc5_26.SKIHMOPFPDC3&ppt=clp&ppn=laptops-store&ssid=j5fwd6niaz0hnchs1645514684273&page={}'
```

```python
stud.format(2)
```

```python
table=[]
from tqdm import tqdm
for i in tqdm(range(1,3)):
    url_all= stud.format(i)
    html_all=requests.get(url_all).text
    soup1=BeautifulSoup(html_all,'html.parser')
    table.append(laptops(soup1))
```

```python
df2=pd.concat(table,axis=0)
df2
```

```python
df2['Price'] = df2['Price'].astype('int64')
df2['Rating'] = df2['Rating'].astype('float64')
df2['Number of ratings'] = df2['Number of ratings'].astype('int64')
df2['Number of reviews'] = df2['Number of reviews'].astype('int64')
df2['Laptop Specifications'] = df2['Laptop Specifications'].tolist()
```
