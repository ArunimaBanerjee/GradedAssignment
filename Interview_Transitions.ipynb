```python
import pandas as pd
def devtomgr(soup):
    emp_name=[i.text.strip().split(',')[0] for i in soup.select('h5.card-title')]
    job_pos=[i.text.strip().split(',')[1].split(' at ')[0] for i in soup.select('h5.card-title')]
    company=[i.text.strip().split(',')[1].split(' at ')[-1] for i in soup.select('h5.card-title')]
    quotes=[i.text.strip().split('.')[0] for i in soup.select('p.card-text')] 
    dt=[i.text.strip().split('\n')[0].split(",'") for i in soup.select('p.card-text')]
    tg=[i.text.strip().split()[4:] for i in soup.select('p.card-text')]
    interview_quotes=quotes[0:10:2]
    date= dt[1:10:2]
    tags=tg[1:10:2]
    return pd.DataFrame({'Employee_Name':emp_name,'Position':job_pos,'Company':company,'Date':date,'Quote':interview_quote, 'Tags':tags}) 
```

```python
devtomgr(soup)
```

```python
stud='https://devtomanager.com/interviews/page/{}/'
import requests
```

```python
tables=[]
from tqdm import tqdm
for i in tqdm(range(1,6)):
    url_all= stud.format(i)
    html_all=requests.get(url_all).text
    soup1=BeautifulSoup(html_all,'html.parser')
    tables.append(devtomgr(soup1))
```

```python
df2=pd.concat(tables,axis=0)
df2
```

```python
comp= np.where(df2['Position']==df2['Company'],np.nan,df2['Company'])
df2['Company']=comp
```
